# Convert XML to CSV

```python
neg_img_dir = '/Users/renatoboemer/signet-ring-cv/data/sig-train-neg'
pos_img_dir = '/Users/renatoboemer/signet-ring-cv/data/sig-train-pos'
annots_dir = '/Users/renatoboemer/signet-ring-cv/data/sig-train-pos/sig-train-pos-xml'
```

```python
class XMLtoCSV:
    def __init__(self, images_dir, annotations_dir):
        self.images_dir = images_dir
        self.annotations_dir = annotations_dir
        self.dataset = self.process_dataset()

    def parse_xml(self, xml_file, image_path):
        """
        parse the XML file and extract bounding box information along with the image path
        """
        tree = ET.parse(xml_file)
        root = tree.getroot()

        boxes = []
        for member in root.findall('object'):
            bbox = member.find('bndbox')
            box_info = {
                'image_path': image_path,
                'xmin': int(bbox.find('xmin').text),
                'ymin': int(bbox.find('ymin').text),
                'xmax': int(bbox.find('xmax').text),
                'ymax': int(bbox.find('ymax').text),
                'label': member.find('name').text
            }
            boxes.append(box_info)
        
        return boxes

    def process_dataset(self):
        """
        process the dataset by pairing each image with its corresponding XML file and creating a df
        """
        rows = []
        for image_file in os.listdir(self.images_dir):
            if image_file.endswith('.jpeg'):
                xml_file = os.path.splitext(image_file)[0] + '.xml'
                xml_path = os.path.join(self.annotations_dir, xml_file)
                image_path = os.path.join(self.images_dir, image_file)

                if os.path.exists(xml_path):
                    boxes = self.parse_xml(xml_path, image_path)
                    rows.extend(boxes)
                else:
                    print(f"annotation for {image_file} not found.")
        
        return pd.DataFrame(rows)

    def get_dataframe(self):
        """
        return the processed dataset as a pandas df
        """
        return self.dataset
```

```python
dataset_processor = XMLtoCSV(pos_img_dir, annots_dir)

pos_df = dataset_processor.get_dataframe()
```

## Check annot image

```python
def show_image_with_boxes(df, image_path):
    """
    display an image with bounding boxes
    """
    # read image
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # filter df for current image
    annotations = df[df['image_path'] == image_path]

    # draw bounding box on the image
    for _, row in annotations.iterrows():
        cv2.rectangle(image, 
                      (int(row['xmin']), int(row['ymin'])), 
                      (int(row['xmax']), int(row['ymax'])), 
                      (0, 255, 0), 2)  # Green box with thickness 2
        cv2.putText(image, 
                    row['label'], 
                    (int(row['xmin']), int(row['ymin']) - 10), 
                    cv2.FONT_HERSHEY_SIMPLEX, 
                    0.5, (0, 255, 0), 2)  # Green text

    # display
    plt.figure(figsize=(10, 10))
    plt.imshow(image)
    plt.axis('off');
```

```python
image_path_to_check = pos_df.iloc[0]['image_path']
show_image_with_boxes(pos_df, image_path_to_check)
```

```python
from PIL import Image

output_dir = '/Users/renatoboemer/signet-ring-cv/data/sig-train-pos-resized'

def resize_images(image_paths, target_size=(512, 512), output_dir=output_dir):
    os.makedirs(output_dir, exist_ok=True)
    resized_paths = []
    for image_path in image_paths:
        with Image.open(image_path) as img:
            img_resized = img.resize(target_size)
            resized_image_path = os.path.join(output_dir, os.path.basename(image_path))
            img_resized.save(resized_image_path)
            resized_paths.append(resized_image_path)
    return resized_paths
```

```python
resized_image_paths = resize_images(image_paths)
```

## Adjust Annotatitions

```python
def adjust_annotations(df, original_size=(1024, 1024), target_size=(512, 512)):
    x_scale = target_size[0] / original_size[0]
    y_scale = target_size[1] / original_size[1]

    df['xmin'] = (df['xmin'] * x_scale).astype(int)
    df['xmax'] = (df['xmax'] * x_scale).astype(int)
    df['ymin'] = (df['ymin'] * y_scale).astype(int)
    df['ymax'] = (df['ymax'] * y_scale).astype(int)
    return df
```

```python
pos_df_adjusted = adjust_annotations(pos_df_resized)

pos_df_adjusted.head()
```
